```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
### LINE CHART ###

# 1) Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# 2) Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# 3) Create the plot
plt.xlabel("Month")
plt.ylabel("Monthly fares")
plt.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="black", linewidth=3, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
### SKILL DRILL - OBJECT ORIENTED APPROACH ###

# 1) Change the line color to green.
# 2) Change the marker to a diamond.
# 3) Change the line width to 2 points.
ax = plt.axes()
ax.plot(x_axis,y_axis, color="green", marker="d", linewidth=2, label='Boston')

# 4) Add a legend for the city of Boston.
ax.legend()
# 5) Add a title and axes labels.
ax.set_xlabel('Date')
ax.set_xlabel('Fare($)')
ax.set_title('PyBer Fare by Month')

# 6) Add a y-axis limit.
ax.set_ylim(5,45)

# 7) Add grid lines.
ax.grid()
```

```python
### BAR CHART ###

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - BAR

plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
### SKILL DRILL - MATLAB APPROACH ###

# 1) Change the bar colors to magenta.
# 2) Add a legend for the city of Boston.

plt.barh(x_axis,y_axis, color='magenta', label='Boston')
plt.legend()

# 3) Add a title and axes labels.
plt.title('PyBer Fare by Month')
plt.xlabel('Date')
plt.ylabel('Fare($)')

# 4) Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

```python
### SKILL DRILL - OBJECT ORIENTED  APPROACH ###

# 1) Creating the plot
# 2) Change the bar color to cyan.
# 3) Add a legend for the city of Chicago.
ax = plt.axes()
ax.barh(x_axis,y_axis, color="cyan", label="Chicago")
ax.legend()

# 4) Add a title and axes labels.
ax.set_title("PybarFare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# 5) Invert the y-axis data.
ax.invert_yaxis()

```

```python
### SCATTER PLOT ### 

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


# 1) Create scatter plot with plt.plot()
#plt.plot(x_axis, y_axis, 'o')

#OR

plt.scatter(x_axis, y_axis)
```

```python
### SKILL DRILL - MATLAB APPROACH ###
# 1) Change the color of the markers to red.
# 2) Add a legend for the city of Chicago.

plt.scatter(y_axis,x_axis,c="r",label="Chicago")
plt.legend()

# 3) Add a title and axes labels.
plt.title("PyBer Fare by month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# 4) Switch the axis so that the Fare($) data is on the x-axis.

# 5) Add a limit for the x-axis data.
plt.xlim(0,45)

# 6) Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()

```

```python
### BUBBLE CHART ###
plt.scatter(x_axis, y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
### Comprehension List ###
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
### SKILL DRILL - OBJECT ORIENTED APPROACH ###

# 1) Change the color of the markers to sky blue.
# 2) Change the size of the markers to 5 times each data point.
# 3) Make the color 20% transparent.
# 4) Add a black edge color to the circles.
# 5) Make the linewidth of the circles 2 points.
# 6) Add a legend for the city of Boston.
# 8) Switch the axis so that the Fare($) data is on the x-axis.

ax = plt.axes()
ax.scatter(y_axis,x_axis, c="skyblue", s=[i*5 for i in y_axis],
            edgecolors='black', alpha=0.9,  linewidth=2, label="Boston")
ax.legend()

# 7) Add a title.
ax.set_title("Pyber")

# 9) Add a limit for the x-axis data.
ax.set_xlim(5,45)

# 10 Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
### PIE CHART ###
# 1) Create pie
#plt.pie(y_axis, labels=x_axis)

# 2) Add labels and explode parameter
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# 3) Increase size of chart:
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# 4) Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
### SKILL DRILL - OBJECT ORIENTED APPROACH ###

# 1) Increase the figure size to 8x8.
plt.subplots(figsize=(8, 8))

# 2) Add a percentage to one decimal place to each wedge of the pie.
ax = plt.axes()

# 3) Explode the two highest percentage months.
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# 4) Add a shadow.
# 5) Add a start angle so that January is at the top.
# 6) Reverse the order so that the month order is in a clockwise direction.
# 7) Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", 
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis,labels=x_axis, autopct='%1.01f%%', explode = explode_values, 
       shadow=True, startangle=92, counterclock=False, colors = colors) 
plt.show()
```
